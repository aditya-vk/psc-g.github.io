```
# @title Imports (run this cell first)
import matplotlib.pyplot as plt
from matplotlib import gridspec
import numpy as np
```

```
# @title Play with the CAFE equations!
# R^ − 1/2 * R* g^ = ((8πG)/c^4) * T^
# Sizes of the dimensions
d1 = 4  # @param {type:'integer'}
d2 = 5  # @param {type:'integer'}
d3 = 6  # @param {type:'integer'}
d4 = 7  # @param {type:'integer'}
R = 10.0  # @param {type:'number'}
save_to_file = True  # @param {type:'boolean'}
G = 6.67408e-11
c = 299792458
# Feel free to change these to something more meaningful!
R_hat = np.reshape(np.cumsum(np.ones((d1, d2, d3, d4))), [d1, d2, d3, d4])
g_hat = np.random.rand(d1, d2, d3, d4) * 20

# Given all the above values, we solve for T^ and plot the different
# dimension combinations.
lhs_multiplier = 8 * np.pi * G / (c ** 4)
T_hat = (R_hat - 0.5 * R * g_hat) / lhs_multiplier
fig = plt.figure(figsize=(32, 32))
gs = gridspec.GridSpec(ncols=3, nrows=3)
for i in range(4):
  for j in range(i + 1, 4):
    tmp_T = np.mean(T_hat, axis=j)
    tmp_T = np.mean(tmp_T, axis=i)
    ax = fig.add_subplot(gs[i, j - 1])
    ax.plot(tmp_T, lw=3)
    ax.set_title('Dimension {} vs {}'.format(i + 1, j + 1), fontsize=22)

if save_to_file:
  # You can download this from the file browser on the left.
  plt.savefig('/tmp/cafe_plots.pdf')
```
